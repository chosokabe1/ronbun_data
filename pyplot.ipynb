```python
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'Times New Roman'
```

```python
n_expansion = [0, 1000, 2000, 3000, 10000]
precision = [0.9342, 0.9292, 0.9136, 0.9282, 0.9334]
recall = [0.8874, 0.8844, 0.8774, 0.8894, 0.9014]
# f_score = [0.9101, 0.9059, 0.8946, 0.9081, 0.9166]
AP5 = [0.9428, 0.9452, 0.9344, 0.9484, 0.9526]
AP595 = [0.6012, 0.6569, 0.6336, 0.6372, 0.663]
```

```python
plt.plot(n_expansion, precision, marker="o", label = "precision")
plt.plot(n_expansion, recall, marker="o", label = "recall")
# plt.plot(n_expansion, f_score, marker="o")
plt.plot(n_expansion, AP5, marker="o", label = "AP0.5")
plt.plot(n_expansion, AP595, marker="o", label = "AP00.5~0.95")

plt.title("result")
plt.xlabel("number of expansion")
plt.legend(loc = "center right")
plt.grid(True)
plt.savefig("result.png")
plt.show()
```
